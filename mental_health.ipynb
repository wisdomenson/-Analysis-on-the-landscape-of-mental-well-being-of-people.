```python
#gender distribution
gender = mh_df['Gender'].value_counts()
gender.sum()
```

```python
#gender distribution
import matplotlib.pyplot as plt 
import seaborn as sbn 

sbn.set_style('whitegrid')
plt.figure(figsize=(6,4))
#mh_df[['Gender']].value_counts().plot(kind='bar', color='g')
gender = mh_df['Gender'].value_counts().plot(kind='bar', color='g')

#for p in gender.patches:
#    gender.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))

for p in gender.patches:
    gender.annotate(str(p.get_height()),(p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5), 
                textcoords='offset points')
    
plt.xticks(rotation=0)
plt.title('Gender Distribution of Respondents')
plt.ylabel('Frequency')
plt.ylim(0,300000)

```

```python
#percentage distribution of gender
import matplotlib.pyplot as plt 
import seaborn as sbn 

#gender = mh_df['Gender'].value_counts().plot.pie(autopct='%1.1f%%')
gender = mh_df['Gender'].value_counts().plot(kind='pie',autopct='%1.1f%%')

plt.legend()
plt.title('Percentage Distribution of Gender')
```

```python
# Combine 5 mental health issues into a DataFrame
import matplotlib.pyplot as plt 
import seaborn as sbn

data = pd.DataFrame({
    'Growing Stress': growing_stress,
    'Changes in Habits': changes_habits,
    'Work Interest': work_interst,
    'Social Weakness': social_weakness,
    'Mental Health Interview': mental_health_intv
})

# Plot each series
for column in data.columns:
    sbn.lineplot(data=data[column], label=column, marker='o')

plt.title('Trends in Various Mental Health Issues')
plt.xlabel('Index')
plt.ylabel('Frequency')
plt.legend(title='Mental Health Issues')
```
